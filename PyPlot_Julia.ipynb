<h2><a id="graficos" style="color: #06386B; font-size: 25px; font-family: 'Open Sans Condensed'; text-decoration:none">Gráficas Simples</a></h2>

<p style="margin: 0; padding:0; color: #343233; text-align: justify; text-justify: inter-word;">En esta seccion presentara el como configurar los componentes de un grafico tales como:</p>
<li>
    <a style="color: #00508f; text-decoration:none">Coordenadas</a>
</li>

<li>
    <a style="color: #00508f; text-decoration:none">Fuentes del título y ejes</a>
</li>

<li>
    <a style="color: #00508f; text-decoration:none">Atributos de un gráfico, tales como color, tamaño, etc.</a>
</li>

````julia
# Gráfico simple sin ninguna configuración
plot(101:200, randn(100));
````

````julia
# Personalizando el color
plot(101:200, randn(100), color="#891838");
````

````julia
# Personalizando el tamaño del gráfico y añadiendo cuadrídulas (grid)
figure(figsize=(7,4));               #(ancho, alto)
plot(101:250, randn(150), color="black");
grid("on", linestyle="--");
````

````julia
figure(figsize=(7,4));
plot(101:250, randn(150), color="#343233", "-o");
grid("on", linestyle="--", color="red");            # Personalizando el color del grid
````

````julia
figure(figsize=(8,5));
plot(101:250, randn(150), color="#008847","-",linewidth=.8);             # Configurando el ancho de la línea del gráfico
title("Distribución Normal", color="red");                # Insertando Título en el gráfico, cambiando el color de letra
xlabel("eje x", color="#00508f", size=18);      # Insertando etiqueta del eje x,cambiando de color y tamaño de la fuente
ylabel("eje y");                                # Insertando etiqueta del eje y
grid("on", linestyle="--");
````

````julia
font1=Dict("family"=>"consolas","color"=>"purple", "size"=>"14");          #Definiendo estilos para el gráfico

figure(figsize=(8,5));
plot(101:250, randn(150), color="#008847", "-", linewidth=.8);
title("Distribución Normal", fontdict=font1);
xlabel("eje x", fontdict=font1); 
ylabel("eje y", fontdict=font1);
grid("on", linestyle="--");
````

````julia
figure(figsize=(8,5));
plot(101:250, randn(150), color="#008847", "-", linewidth=.8);
title(L"$\mathcal{N}(\mu,\,\sigma^{2})$");                           # tipografía estilo LaTeX
xlabel(L"$eje x$"); 
ylabel(L"$eje y$");
grid("on", linestyle="--");
````

````julia
# Cambiando de color los bordes del gráfico

figure(figsize=(7,4));
ax=axes();
plot(101:250, randn(150), color="#008847", "-o");
ax[:spines]["top"][:set_color]("red")
ax[:spines]["right"][:set_color]("blue")
ax[:spines]["left"][:set_color]("orange")
ax[:spines]["bottom"][:set_color]("green")
grid("on", linestyle="--");
````

````julia
figure(figsize=(7,4));
ax=axes();
plot(101:250, randn(150), color="#00508f", "-o");
ax[:tick_params](axis="y", colors="red");          #Color del eje y
ax[:tick_params](axis="x", colors="blue");         #Color del eje x
grid("on", linestyle="--");
````

````julia
figure(figsize=(7,4));
ax=axes();
plot(101:250, randn(150),"o-",markerfacecolor="red", markeredgecolor="blue", color="black");
grid("on", linestyle="--");
````

````julia
figure(figsize=(7,4));
ax=axes();
plot(101:250, randn(150), marker="v",markerfacecolor="red", markeredgecolor="blue", color="black", markersize=12);
grid("on", linestyle="--");
````

````julia
figure(figsize=(7,4));
ax=axes();
plot(101:250, randn(150), marker="d",markerfacecolor="#209281", markeredgecolor="red", color="blue", markersize=8);
ax[:spines]["top"][:set_visible](false);
ax[:spines]["right"][:set_visible](false);
ax[:spines]["left"][:set_visible](false);
ax[:spines]["bottom"][:set_visible](false);
grid("on", linestyle="--");
````

<h2><a id="pie" style="color: #06386B; font-size: 25px; font-family: 'Open Sans Condensed'; text-decoration:none">Pie Plot</a></h2>

````julia
N = 20
theta = linspace(0.0, 2*pi, N)
radii = 10*rand(N)
width = pi./4*rand(N)

fig = figure("pyplot_piechart",figsize=(4, 4))
ax = plt[:subplot](111, projection="polar")
bars = ax[:bar](theta, radii, width=width, bottom=0.0)

for (r, bar) in zip(radii, bars)
    bar[:set_facecolor](plt[:cm][:viridis](r/10))
    bar[:set_alpha](0.8)
end
````

````julia
N = 150
r = 2*rand(N)
theta = 2*pi*rand(N)
area = 200 * r.^2
colors = theta

fig = figure("pyplot_piechart",figsize=(4, 4));
ax = plt[:subplot](111, projection="polar");
c = ax[:scatter](theta, r, c=colors, s=area, cmap="hsv", alpha=0.75);
````
